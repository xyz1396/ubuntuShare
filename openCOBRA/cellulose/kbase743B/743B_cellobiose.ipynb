```python
import cobra.test
import cobra.io
import numpy as np
import pandas as pd
import math
```

```python
cellobiose=cobra.io.read_sbml_model("743B_cellobiose.xml")
display(cellobiose)
```

### medium

```python
def getMediumContent(model):
    df=pd.DataFrame()
    for i in model.medium.keys():
        compound=model.metabolites.get_by_id(i[3:])
        df[i]=pd.Series({'Metabolite identifier': compound.id, 'Name':compound.name, \
            'Formula':compound.formula}) 
    return df.T
getMediumContent(cellobiose)
```

### solution

```python
solution=cellobiose.optimize()
summary=cellobiose.summary(solution,0.00001).to_frame()
j=0
for i in summary['IN_FLUXES']['ID']:
    summary.iloc[j,0]=cellobiose.metabolites.get_by_id(i).name
    j=j+1
j=0
for i in list(summary['OUT_FLUXES']['ID']):
    if i==i:
        summary.iloc[j,2]=cellobiose.metabolites.get_by_id(i).name
        j=j+1
```
