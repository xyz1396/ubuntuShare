```python
import cobra.test
import cobra.io
import pandas as pd
```

```python
%%capture
iMR=cobra.io.read_sbml_model("iMR539.xml")
```

```python
df=pd.read_csv("S2 to GP6.csv")
S2=list(df.S2)
GP6=list(df.GP6)
for i,gene in enumerate(iMR.genes):
    id = gene.id
    if id in S2:
        iMR.genes[i].id=GP6[S2.index(id)]
```
